```python
%pip install katna
from Katna.video import Video
from Katna.writer import KeyFrameDiskWriter

def extract(video_file):
    
    vd = Video()

    # number of images to be returned
    no_of_frames_to_returned = 2

    # initialize diskwriter to save data at desired location
    diskwriter = KeyFrameDiskWriter(location="./2_extracted_keyframe")

    # Video file path
    video_file_path = f'./Activity_RGBvideo_dataset/a/b/{video_file}' 

    # extract keyframes and process data with diskwriter
    vd.extract_video_keyframes(
        no_of_frames=no_of_frames_to_returned, file_path=video_file_path,
            writer=diskwriter)
    
```

```python
from collections import defaultdict

def filter_files_with_pairs(file_list):
    # Create a dictionary to count occurrences of the base names
    file_dict = defaultdict(int)
    
    # Count the occurrences of each base name
    for file in file_list:
        base_name = file.rsplit('_', 1)[0]
        file_dict[base_name] += 1
    
    # Filter and return only the files with a pair
    result = [file for file in file_list if file_dict[file.rsplit('_', 1)[0]] == 2]
    
    return result



images = filter_files_with_pairs(images)
```
